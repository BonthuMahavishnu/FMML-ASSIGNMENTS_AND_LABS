## Using learnt representations
In practice, very few people train an entire Convolutional Network from scratch (with random initialization), because it is relatively rare to have a dataset of sufficient size. Instead, it is common to pretrain a ConvNet on a very large dataset (e.g. ImageNet, which contains 1.2 million images with 1000 categories), and then use the ConvNet either as an initialization or a fixed feature extractor for the task of interest.


We'll train a model to classify ants and bees. We have about 120 training images each for ants and bees. There are 75 validation images for each class. Usually, this is a very small dataset to generalize upon, if trained from scratch. Since we are using transfer learning, we should be able to generalize reasonably well.

This dataset is a very small subset of imagenet.




```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, SubsetRandomSampler
import torch.nn as nn
from torch import optim
import torch.nn.functional as F

import matplotlib.pyplot as plt
import numpy as np
import time
import os
```

```python
# Device configuration (whether to run on GPU or CPU)
device = 'cpu' #torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Download and extract dataset
!wget https://download.pytorch.org/tutorial/hymenoptera_data.zip
!unzip -q hymenoptera_data.zip
```

```python
# Data augmentation and normalization for training
# Just normalization for validation
train_transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
val_transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

data_dir = './hymenoptera_data'
train_dataset = ImageFolder(os.path.join(data_dir, 'train'), train_transform)
val_dataset = ImageFolder(os.path.join(data_dir, 'val'), val_transform)
train_dataloader = DataLoader(train_dataset, batch_size=4,shuffle=True, num_workers=2)
val_dataloader = DataLoader(val_dataset, batch_size=4,shuffle=True, num_workers=2)
class_names = train_dataset.classes
```

```python
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(train_dataloader))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

```python
def train_model(model, train_dataloader, criterion, optimizer, num_epochs=25):

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}')
        print('-' * 10)
        model.train()
        running_loss = 0.0
        running_corrects = 0

        # Iterate over data.
        for i, (inputs, labels) in enumerate(train_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward
            # track history if only in train
            with torch.set_grad_enabled(True):
              outputs = model(inputs)
              _, preds = torch.max(outputs, 1)
              loss = criterion(outputs, labels)

              # backward + optimize only if in training phase
              loss.backward()
              optimizer.step()

            # statistics
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(train_dataloader.dataset)
        epoch_acc = running_corrects.double() / len(train_dataloader.dataset)

        print(f'Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

        print()
```

```python
def visualize_model(model, num_images=10):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(val_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title(f'predicted: {class_names[preds[j]]}')
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
from torchvision import models

model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features
# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
model_ft.fc = nn.Linear(num_ftrs, 2)

model_ft = model_ft.to(device)

loss_func = nn.CrossEntropyLoss()

# Observe that all parameters are being optimized
optimizer = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 5
train_model(model_ft, train_dataloader, loss_func, optimizer, num_epochs)
```

```python
visualize_model(model_ft)
```

#1.
**Data augmentations like resizing, cropping, flipping, rotating, and adding noise serve several significant purposes when applied to training data:**

***Increased Dataset Size:***

  By applying various augmentations to the original dataset, you effectively increase the size of your training dataset. This larger dataset provides more diverse examples for the model to learn from, which can help improve its generalization and robustness.

***Improved Generalization:***

Augmentations introduce variations in the training data, making the model less sensitive to specific variations in the input during training. This helps prevent overfitting by exposing the model to a broader range of scenarios that it might encounter during inference.

***Translation Invariance:***

 Techniques like translation (e.g., cropping and padding) help the model become invariant to translations in the input data. This means that the model can still recognize objects even if they appear in different positions within the image.

***Rotation and Scale Invariance:***

 Rotating and scaling images can help the model become invariant to rotations and changes in scale, making it more robust to variations in viewpoint and object size.

***Regularization:***

 Augmentations act as a form of regularization by introducing noise and perturbations to the training data. This can help prevent the model from memorizing the training examples and instead encourages it to learn more robust and generalizable features.

***Balancing Classes:***

 Augmentations can also be used to balance class distributions in the dataset. For example, if certain classes are underrepresented, you can apply augmentation techniques to increase the number of samples for those classes, helping to address class imbalance issues.

***Training Duration:***

 Training a deep neural network like ResNet from scratch can be computationally expensive and time-consuming, especially without the benefit of transfer learning from pretrained weights.

***Overfitting Risk:***

 Without the regularization effect of pretrained weights, there's a higher risk of overfitting, especially if the dataset is small or lacks diversity.

***Data Requirements:***

 Training from scratch often requires a larger and more diverse dataset to learn meaningful representations effectively.

***Resource Intensiveness:***

 Training deep neural networks from scratch demands significant computational resources, including GPU capabilities and memory

***Optimization Challenges:***

 Training from scratch might require careful tuning of hyperparameters, learning rates, and regularization techniques to achieve good performance.

The performance of a pretrained ResNet model on a task like classifying ants vs. bees images would depend on several factors:

***Domain Similarity:***

 If the dataset on which the ResNet model was pretrained is significantly different from the ants vs. bees dataset, the pretrained model might not capture relevant features for the new task. For example, if the pretrained dataset is related to medical imaging, the features learned by the model may not be directly applicable to insect classification.

***Task Similarity:***

 Even if the datasets are different, if the underlying tasks share similar characteristics, the pretrained model might still provide useful feature representations. For instance, if both datasets involve image classification tasks with similar object categories (e.g., animals), the pretrained model might learn general features that are transferable to the new task.

***Fine-Tuning:***

 Fine-tuning the pretrained model on the ants vs. bees dataset can help adapt the learned features to the new task. By updating the weights of the pretrained model during training on the new dataset, the model can learn to extract task-specific features while retaining useful generalizations from the pretrained weights.

***Amount of Data:***

 The amount and diversity of data in the ants vs. bees dataset also play a crucial role. If the dataset is small or significantly different from the pretrained dataset, fine-tuning may be less effective, and the model may struggle to generalize well to the new task.

***Evaluation:***

 Ultimately, the performance of the pretrained model on the ants vs. bees dataset should be evaluated empirically through validation and testing. Experimenting with both fine-tuning and training from scratch can help determine the best approach for achieving optimal performance on the specific task and dataset.
